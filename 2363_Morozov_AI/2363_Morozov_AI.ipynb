## Основные теоретические положения

Нормальное распределение двумерного случайного вектора определяется функцией плотности:
$p(\vec{x}) = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \exp\left( -\frac{1}{2} \cdot (\vec{x} - \vec{\mu})^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \vec{\mu}) \right)$
где:\

$\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}$ - Ковариационная матрица;

$\vec{x} = \begin{pmatrix} \xi \\ \eta \end{pmatrix};$\
$\vec{\mu} = \mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix}.$


Афинное преобразование:

$\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$ — исходный случайный вектор;\
$\vec{\xi}_{st} \sim \mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ — стандартный нормальный случайный вектор;\
$\vec{\xi}_{st} = A \cdot \vec{\xi} + b$ — афинное преобразование случайного вектора;\
$\Sigma_{st} = A \cdot \Sigma \cdot A^{\intercal}$ — получение стандартной нормальной ковариационной матрицы из исходной.\
$\quad$

Ортогональное преобразование, которое переводит центрированный вектор в вектор с независимыми компонентами, можно описать следующим образом:

$Q^{\intercal}$ — это матрица, составленная из собственных векторов матрицы ковариации $\Sigma;$\
При вычислении выражения $Q \cdot \Sigma \cdot Q^{\intercal}$ должна получится матрица $D$ вида: $\begin{pmatrix} \ldots & 0 \\ 0 & \ldots \end{pmatrix};$\
Матрица $D$ является диагональной матрицей из собственных значений $\Sigma^{-1},$\
при этом $\Sigma = Q \cdot D^{-1} \cdot Q^{\intercal},$\
Плотность распределения независимых компонентов центрированного вектора будет иметь вид:\
$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \cdot \exp\left(-\frac{1}{2} \cdot \left[  \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$\
$\quad$

### Плотность двумерного нормального распределения имеет вид: $p_{\xi,\eta} = C \cdot exp\left(-\frac{1}{2} (2x^2 - 4xy + 5y^2 - 12x + 18y + 21) \right)$;

#### 1) Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi,\eta}$

$q(x,y) = 2x^2 - 4xy + 5y^2 - 12x + 18y + 21 = 2(x^2 - 2xy + y^2) - 12(x - y) + 3y^2 + 6y + 21 = 2((x - y)^2 - 6(x - y) + 9) + 3y^2 + 6y + 3 = 2((x - y) - 3)^2 + 3(y^2 + 2y +1) = 2((x - 2) - (y + 1))^2 + 3(y + 1)^2 = 2(x - 2)^2 - 4(x - 2)(y + 1) + 2(y + 1)^2 + 3(y + 1)^2$

Вектор математического ожидания:

$\mathbb{E}\begin{pmatrix}\xi \\ η \end{pmatrix} = \begin{pmatrix}2 \\ -1 \end{pmatrix}$

Ковариационная матрица $Σ$:

$Σ^{-1} = \begin{pmatrix} 2 & -2 \\ -2 & 5\end{pmatrix}$

$\begin{vmatrix} Σ^{-1} \end{vmatrix} = 6; \frac{1}{\begin{vmatrix} Σ^{-1} \end{vmatrix}} = \frac{1}{6}$

$Σ = \frac{1}{6} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2\end{pmatrix}$

Дисперсия:

$\mathbb{D}\xi = 5 \cdot \frac{1}{6} = \frac{5}{6}$

$\mathbb{D}\eta = 2 \cdot \frac{1}{6} = \frac{1}{3}$

Ковариация:

$\mathop{\mathrm{cov}}\limits(\xi, \eta) = \mathop{\mathrm{cov}}\limits(\eta, \xi) = 2 \cdot \frac{1}{6} = \frac{1}{3}$

Коэффициент корреляции:

$\rho(\xi, \eta) = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi \cdot \mathbb{D}\eta}} = \frac{\frac{1}{3}}{\sqrt{\frac{5}{6}\cdot\frac{1}{3}}} = \frac{\sqrt{10}}{5}$

Константа $C$:

$C = \frac{1}{2\pi\sqrt{\frac{1}{\begin{vmatrix} Σ^{-1} \end{vmatrix}}}} = \frac{1}{2π\sqrt{\frac{1}{6}}} = \frac{\sqrt{6}}{2\pi}$

График плотности $p_{\xi,\eta}$:

```r
density_func <- function(x, y) {
  C <- sqrt(6) / (2 * pi)
  exponent <- -0.5 * ( 2 * (x - 2)^2 - 4 * (x - 2) * (y + 1) + 2 * (y + 1)^2 + 3 * (y + 1)^2)
  return(C * exp(exponent))
}

x <- seq(-5, 5, length.out = 100)
y <- seq(-5, 5, length.out = 100)

density_matrix <- outer(x, y, density_func)
persp(x, y, density_matrix,
        xlab = "X-axis",
        ylab = "Y-axis",
        zlab = "Density",
        main = "График плотности",
        phi = 30, theta = 30)
```

$q(x,y) = 2((x - y) - 3)^2 + 3(y^2 + 2y +1) = (\sqrt{2}x - \sqrt{2}y - 3\sqrt{2})^2 + (\sqrt{3}y + \sqrt{3})^2$

Аффинное преобразование:

$\begin{pmatrix} ξ_{st} \\ η_{st} \end{pmatrix} = A \cdot \begin{pmatrix} ξ \\ η \end{pmatrix} + \begin{pmatrix} α \\ β \end{pmatrix}$

$\mathbb{E}\begin{pmatrix} ξ_{st} \\ η_{st} \end{pmatrix} = \begin{pmatrix} \sqrt{2} & -\sqrt{2} \\ 0 & \sqrt{3}\end{pmatrix} \cdot \begin{pmatrix} 2 \\ -1 \end{pmatrix} + \begin{pmatrix} -3\sqrt{2} \\ \sqrt{3} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

$Σ_{st}=\frac{1}{6} \cdot  \begin{pmatrix} \sqrt{2} & -\sqrt{2} \\ 0 & \sqrt{3}\end{pmatrix} \begin{pmatrix} 5 & 2 \\ 2 & 2\end{pmatrix}\begin{pmatrix} \sqrt{2} & 0 \\ -\sqrt{2} & \sqrt{3}\end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$

#### 3) Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$Σ^{-1} = \begin{pmatrix} 2 & -2 \\ -2 & 5\end{pmatrix}$

$\begin{vmatrix} Σ^{-1}-λE\end{vmatrix} = 0$

$\begin{vmatrix} \begin{pmatrix} 2 & -2 \\ -2 & 5\end{pmatrix} - \begin{pmatrix} λ & 0 \\ 0 & λ\end{pmatrix} \end{vmatrix} = \begin{vmatrix}\begin{pmatrix}2-λ & -2 \\ -2 & 5-λ\end{pmatrix}\end{vmatrix} = 0$

$(2-λ)(5-λ)-4=0$

$\begin{cases}λ_1 = 1\\ λ_2 = 6\end{cases}$

$\begin{cases}n_1-2m_1=0 \\ -2n_1+4m_1=0\end{cases}$

$\begin{cases}n_1=2m_1\\ m_1 = \alpha, \alpha∈\mathbb{R}\end{cases}$

$\alpha = \sqrt{2^2+1^2} = \sqrt{5}$

$v_1 = \frac{1}{\sqrt{5}} \cdot \begin{pmatrix} 2 \\ 1 \end{pmatrix}$

$\begin{cases}-4n_2-2m_2=0 \\ -2n_2-m_2=0\end{cases}$

$\begin{cases}-2n_2=m_2 \\ m_2 = \beta, \beta \in \mathbb{R}\end{cases}$

$\beta = \sqrt{(-\frac{1}{2})^2+1^2} = \sqrt{\frac{5}{4}}$

$v_2 = \frac{2}{\sqrt{5}} \cdot \begin{pmatrix} -\frac{1}{2} \\ 1 \end{pmatrix}$

$Q^{\intercal} = \frac{1}{\sqrt{10}}\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} = Q$

$Q\cdotΣ^{-1}\cdot Q^{\intercal} = \frac{1}{\sqrt{5}}\begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix} \cdot \begin{pmatrix} 5 & 2 \\ 2 & 2\end{pmatrix} \frac{1}{\sqrt{5}}\begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} = \begin{pmatrix} 6 & 0 \\ 0 & 1\end{pmatrix} = A$

$p_{\xi,\eta}(x', y') = \frac{1}{2\pi\sqrt{\begin{vmatrix}\Sigma\end{vmatrix}}} \cdot \exp\left(-\frac{1}{2} \cdot \left(  \begin{pmatrix} x' & y' \end{pmatrix} \cdot A \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right)\right) = \frac{\sqrt{6}}{2\pi} \cdot\exp\left(-\frac{1}{2} \cdot ( 6x'^2 + 1y'^2)\right)$

```r
density_function <- function(x, y) {
  (sqrt(6) / (2 * pi)) * exp(-0.5 * (6 * x^2 + 1 * y^2))
}

# Generate a grid of x and y values
x_vals <- seq(-3, 3, length.out = 100)
y_vals <- seq(-3, 3, length.out = 100)
z_vals <- outer(x_vals, y_vals, density_function)

# Create a 3D surface plot
persp(x_vals, y_vals, z_vals, col = "lightblue",
        xlab = "X'", ylab = "Y'", zlab = "Density",
        main = "3D Density Distribution", phi = 30, theta = 30)
```

#### 4) Вычислить характеристики совместного распределения случайного вектора $(4ξ+2η, 2ξ-η)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.

$\vec{v} = A \cdot \begin{pmatrix} \xi \\ η \end{pmatrix} = (4ξ+2η, 2ξ-η)^{\intercal}$

Таким образом $A = \begin{pmatrix} 4 & 2 \\ 2 & -1 \end{pmatrix}$

Обозначим $A \cdot \vec{\mu} = \mathbb{E}[\vec{v}] = \vec{\mu}{\vec{v}}$ и $A \cdot \Sigma \cdot A^{\intercal} = \Sigma{\vec{v}}$

```r
mu <- matrix(c(2, -1), ncol = 1)
Sigma <- matrix(c(5/6, 1/3, 1/3, 1/3), nrow = 2)
A <- matrix(c(4, 2, 2, -1), nrow = 2)

mu_v <- A %*% mu
Sigma_v <- A %*% Sigma %*% t(A)

det_Sigma_v <- det(Sigma_v)

cat("mu_v:","\n")
prmatrix(mu_v)

cat("Sigma_v ⋅ 6:","\n")
prmatrix(Sigma_v * 6)

cat("det_Sigma_v ⋅ 6:", det_Sigma_v * 6)
```

$Σ_{\vec{v}} = \frac{4}{75} \cdot \begin{pmatrix} 120 & 36 \\ 36 & 14 \end{pmatrix}$

$\vec{\mu}_\vec{v} = \begin{pmatrix} 6 \\ 5 \end{pmatrix}$


$Σ_{\vec{v}}^{-1} = \frac{3}{32} \cdot \begin{pmatrix} 14 & -36 \\ -36 & 120 \end{pmatrix}$

Зная данные элементы восстановим функцию плотности:
$𝛲_{\vec{v}}(\vec{t}) = \frac{2\sqrt{6}}{3\pi} ⋅ \exp\left(-\frac{1}{2} \cdot \frac{3}{32}\left(14\left(ξ - 6\right)^2 -72\left(ξ - 6\right)(η - 5) + 120(η - 5)^2\right)\right)$

График плотности полученного распределения:

```r
P_z <- function(xi, eta) {
  2 * sqrt(6) / (3 * pi) * exp(-0.5 * (3/32) * (14 * (xi - 6)^2 - 72 * (xi - 6) * (eta - 5) + 120 * (eta - 5)^2))
}

xi <- seq(0, 15, length = 50)
eta <- seq(-7, 8, length = 50)
z <- outer(xi, eta, P_z)

persp(xi, eta, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 0, phi = 0)
persp(xi, eta, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 90, phi = 90)
persp(xi, eta, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 45, phi = 45)
```
