## Основные теоретические положения

Математическое ожидание:

$\mu = \mathbb E \vec{\xi} = (\mathbb E \xi1, ..., \mathbb E \xi_n)^{\intercal}$

Матрица ковариаций:

$\Sigma = \begin{pmatrix}
\mathbb{D}_\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta)\\
\mathop{\mathrm{cov}}\nolimits(\xi, \eta) & \mathbb{D}_\eta
\end{pmatrix}$

Нормальное распределение (одномерный):

$ \xi \sim \mathcal{N}(a, \sigma^2) $

$ p_\xi (x) = \cfrac{1}{\sigma\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{x-a}{\sigma}\right)^2\right) $

1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $ \rho_{\xi,\eta} $.
   
$ q(x, y)=3x^2+3xy+7y^2+9x+17y+13=3(x+\frac{y}{2})^2+\frac{25}{4}y^2+9x+17y+13=3((x+\frac{y}{2})+\frac{3}{2})^2+\frac{25}{4}y^2+12.5y+6.25=3((x+\frac{y}{2})+\frac{3}{2})^2+25(\frac{y}{2}+\frac{1}{2})^2= $
$ =3(x+1)^2+6(x+1)(\frac{y}{2}+\frac{1}{2})+3(\frac{y}{2}+\frac{1}{2})^2+25(\frac{y}{2}+\frac{1}{2})^2=3(x+1)^2+2\cdot1.5(x+1)(y+1)+7(y+1)^2$

Вектор математических ожиданий и ковариационные характеристики:

$\mathbb{E}\begin{pmatrix}\xi\\\eta\end{pmatrix}=\begin{pmatrix}-1\\-1\end{pmatrix}, \Sigma^{-1}=\begin{pmatrix}3&\frac{3}{2}\\\frac{3}{2}&7\end{pmatrix}\Rightarrow\Sigma=\frac{1}{18.75}\begin{pmatrix}7&-\frac{3}{2}\\-\frac{3}{2}&3\end{pmatrix}, \mathbb{D}\xi=\frac{7}{18.75}, \mathbb{D}\eta=\frac{3}{18.75}, \mathop{\mathrm{cov}}\nolimits(\xi, \eta)=-\frac{2}{25}, \rho(\xi, \eta)=-\frac{2\sqrt{21}}{7}, C=\frac{1}{2\pi\sqrt{\frac{1}{18.75}}}$

График плотности:


```python
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
x = np.linspace(-8, 8, 400)
y = np.linspace(-8, 8, 400)
X, Y = np.meshgrid(x, y)
С = (1/(2*np.pi*np.sqrt(1/18.75)))
Z = С*np.exp(-0.5*(3*X**2+3*X*Y+7*Y**2+9*X+17*Y+13))
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
```
