## Основные теоретические положения

Нормальное распределение двумерного случайного вектора определяется функцией плотности:
$p(\vec{x}) = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \exp\left( -\frac{1}{2} \cdot (\vec{x} - \vec{\mu})^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \vec{\mu}) \right)$
где:\
$\vec{x} = \begin{pmatrix} \xi \\ \eta \end{pmatrix},$
$\vec{\mu} = \mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix},$
$\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}.$\
$\quad$\
Ортогональное преобразование, переводящее соответствующий центрированный вектор в вектор с независимыми компонентами:\
$Q^{\intercal}$ — матрица из собственных векторов матрицы $\Sigma;$\
При вычислении выражения $Q \cdot \Sigma \cdot Q^{\intercal}$ должна получится матрица $D$, которая будет являться диагональной матрицей из собственных значений $\Sigma^{-1}$.\
Плотность распределения независимых компонентов центрированного вектора будет иметь вид:\
$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \cdot \exp\left(-\frac{1}{2} \cdot \left[  \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$\
$\quad$\
Для многомерного нормального распределения условное распределение также является нормальным.\
Условное распределение $\xi | \eta = y$ имеет следующую форму:\
$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$, где $\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{\sigma_{\xi\eta}}{\sigma_{\eta}^2} \cdot (y - \mu_{\eta}), \mathbb{D}[\xi|\eta] = \sigma_{\xi}^2 - \frac{\sigma_{\xi\eta}^2}{\sigma_{\eta}^2}.$

4. Вычислить характеристики совместного распределения случайного вектора $(3\xi-\eta,\xi+4\eta)^T$ и записать его плотность. Построить график плотности полученного вектора.

Пусть $r = 3\xi-\eta$ и $k = \xi+4\eta$

(По св-ву линейности математического ожидания)\
$\mathbb{E}(3\xi-\eta) = 3\mathbb{E}\xi-\mathbb{E}\eta = 3 \cdot 1 - (-1) = 4$\
$\mathbb{E}(\xi+4\eta) = \mathbb{E}\xi+4\mathbb{E}\eta = 1 + 4 \cdot(-1) = -3$

$\mathbb{D}(3\xi-\eta) = 9\mathbb{D}\xi + \mathbb{D}\eta = 9 \cdot \frac{4}{25} + \frac{28}{75} = \frac{136}{75}$\
$\mathbb{D}(\xi+4\eta) = \mathbb{D}\xi + 16\mathbb{D}\eta = \frac{4}{25} + 16\cdot\frac{28}{75} = \frac{460}{75}$

$cov(3\xi-\eta,\xi+4\eta) = \mathbb{E}((3\xi-\eta)\cdot(\xi+4\eta)) - \mathbb{E}(3\xi-\eta)\cdot\mathbb{E}(\xi+4\eta) = \mathbb{E}(3\xi^2+11\xi\eta-4\eta^2) - \mathbb{E}(3\xi-\eta)\cdot\mathbb{E}(\xi+4\eta) =$

$=3\mathbb{E}\xi^2+11\mathbb{E}\xi\eta-4\mathbb{E}\eta^2-(3(\mathbb{E}\xi)^2+11\mathbb{E}\xi\mathbb{E}\eta-4(\mathbb{E}\eta)^2) = 3\mathbb{D}\xi-4\mathbb{D}\eta = 3 \cdot \frac{4}{25} - 4 \cdot \frac{28}{75} = -\frac{76}{75}$ Так как величины незасивимы

Тогда ковариационная матрица: 
$\Sigma_{XY} =\frac{1}{75}\begin{pmatrix}
136 & -76  \\
-76  & 460
\end{pmatrix}
$

$\rho(r, k) = \frac{\mathop{\mathrm{cov}}\nolimits(r, k) }{\sqrt{\mathbb{D}_r \mathbb{D}_k}}= -\frac{76}{75}\cdot\frac{75}{\sqrt{136\cdot460}} = -\frac{19}{\sqrt{3910}}$

$P_{r,k}(x,y) = \frac{1}{2\pi\sqrt{\mathbb{D}(x)\mathbb{D}(y)}\sqrt{1-\rho(\xi, \eta)^2}}\cdot exp(-\frac{1}{2(1-\rho(\xi, \eta)^2)}(\frac{(x-\mathbb{E}(x))^2}{\mathbb{D}(x)}-\frac{2\rho(\xi, \eta)(x-\mathbb{E}(x))(y-\mathbb{E}(y))}{\sqrt{\mathbb{D}(x)\mathbb{D}(y)}}+\frac{(y-\mathbb{E}(y))^2}{\mathbb{D}(y)})) = $

$=\frac{75}{104\pi\sqrt{82110}}\cdot exp(-\frac{3910}{7098}(\frac{75(x-4)^2}{136}-\frac{285}{3128}\cdot(x-4)(y+3)+\frac{75(y+3)^2}{460}))=$

$=\frac{75}{104\pi\sqrt{82110}}\cdot exp(-\frac{3910}{7098}\cdot\frac{1}{3128}(1725x^2-14655x-285xy+35610+4200y+510y^2))$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def density_function(x, y):
    const = 75  / (104 * np.pi* np.sqrt(82110))
    exp = -(3910/(7098*3128)) * (1725 * x**2 -14655*x-285*x*y+35610+4200*y + 510 * y**2)
    return const * np.exp(exp)

x_values = np.linspace(-12, 18, 1000)
y_values = np.linspace(-15, 15, 1000)

X, Y = np.meshgrid(x_values, y_values)

Z = density_function(X, Y)
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(X, Y, Z, cmap='Greys', edgecolor='none')

ax.set_xlabel("x'")
ax.set_ylabel("y'")
ax.set_zlabel('Плотность')
ax.set_title("Плотность распределения независимых компонент вектора")

plt.show()
```
