## Основные теоретические положения

Нормальное распределение (одномерный случай):

$ \xi \sim \mathcal{N}(a, \sigma^2) $

$ p_\xi (x) = \cfrac{1}{\sigma\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{x-a}{\sigma}\right)^2\right) $

Многомерное нормальное распределение:

$ \overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma) $

$ p_{\overrightarrow{\xi}} \left(\overrightarrow{x}\right) = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} \cdot \exp \left(-\cfrac{1}{2}\left(\overrightarrow{x}-\mu\right)^{\intercal}\Sigma^{-1}\left(\overrightarrow{x}-\mu\right)\right) $


Матрица ковариаций:

$\Sigma = \begin{pmatrix}
\mathbb{D}_\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta)\\
\mathop{\mathrm{cov}}\nolimits(\xi, \eta) & \mathbb{D}_\eta
\end{pmatrix}$

Математическое ожидание:

$\mu = \mathbb E \vec{\xi} = (\mathbb E \xi1, ..., \mathbb E \xi_n)^{\intercal}$


СКО:

$\sigma_{ij} = \mathop{cov}(\xi_i, \xi_j)$

Преобразование характеристик:

$ \overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma) $

$ \overrightarrow{\eta} = A \overrightarrow{\xi} + \overrightarrow{b} $

$ \overrightarrow{\eta} \sim \mathcal{N}(A \overrightarrow{\mu} + \overrightarrow{b}, A \Sigma A^{\intercal}) $

Условное распределение для абсолютно непрерывных величин:

$ p_{\xi|\eta=y_0} (x) = \cfrac{p_{\xi, \eta}  (x, y_0)}{p_\eta  (y_0)} $

Стандартное нормально распределение - нормальное распределение с нулевым мат. ожиданием и единичной дисперсией. В многомерном случае - нулевой вектор мат. ожидания и единичная ковариационная матрица.

Центрированный случайный вектор - случайный вектор с нулевым мат. ожиданием.


Аффинное преобразование - преобразование вида:

$ \begin{pmatrix}
x\\
y
\end{pmatrix} = A \begin{pmatrix}
u\\
v
\end{pmatrix} + \begin{pmatrix}
\alpha\\
\beta
\end{pmatrix} $

###1. Вычислим вектор математических ождианий и ковариационные характеристики данного случайного вектора. Для начала приведем квадратичную форму $Q(x, y)$ к каноническому виду:
$Q(x, y) = 4x^2 - 9xy + 16y^2 + 7x + 14y + 14 =$\
$= 4x^2-9xy + \frac{81}{16}y^2 + 7x + 14y + 14 + \frac{175}{16}y^2 =$\
$= 4(x^2 - \frac{9}{4}xy + (\frac{9}{8}y)^2) + 7x - \frac{63}{8}y + 14 + \frac{175}{16}y^2 + \frac{175}{8}y =$\
$= 4(x - \frac{9}{8}y)^2 + 7(x - \frac{9}{8}y) + 14 + \frac{175}{16}y^2 + \frac{175}{8}y =$\
$= 4((x - \frac{9}{8}y)^2 + \frac{7}{4}(x - \frac{9}{8}y)) + \frac{49}{16} + \frac{175}{16}(y+1)^2 =$\
$= 4((x - \frac{9}{8}y)^2 + 2(x - \frac{9}{8}y)\frac{7}{8} + (\frac{7}{8})^2) + \frac{175}{16}(y+1)^2 =$\
$=(*) 4((x - \frac{9}{8}y) + \frac{7}{8} )^2 + \frac{175}{16}(y+1)^2 (*)=$\
$= 4( (x+2) - \frac{9}{8}(y + 1))^2 + \frac{175}{16}(y + 1)^2 =$\
$= 4( (x+2)^2  - \frac{9}{4}(x + 2)(y + 1) + \frac{81}{64} (y + 1)^2) + \frac{175}{16}(y+1)^2 =$\
$= 4(x + 2)^2  - 9(x + 2)(y + 1) + \frac{81}{16} (y + 1)^2 + \frac{175}{16}(y+1)^2 =$\
$= 4(x + 2)^2  - 9(x + 2)(y + 1) + 16(y + 1)^2 $\
Получаем:\
$p_{\xi,\eta}(x, y) = C \cdot \exp\left(-\frac{1}{2} \cdot \left[4(x + 2)^2  - 9(x + 2)(y + 1) + 16(y + 1)^2\right]\right).$\
Вектор математических ожиданий данного случайного вектора имеет вид:\
$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -2 \\ -1 \end{pmatrix}.$\
Получим матрицу $\Sigma^{-1}:$\
$\Sigma^{-1} = \begin{pmatrix} 4 & -4.5 \\ -4.5 & 16 \end{pmatrix}.$\
Найдем ковариационную матрицу $\Sigma:$\
$\Sigma = \frac{1}{|\Sigma^{-1}|} \cdot \begin{pmatrix} 16 & 4.5 \\ 4.5 & 4 \end{pmatrix} = \frac{1}{43.75} \cdot \begin{pmatrix} 16 & 4.5 \\ 4.5 & 4 \end{pmatrix}$

Ковариционная матрица имеет вид: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix},$ следовательно запишем ковариационные характеристики:\
$\mathbb{D}\xi = \frac{1}{43.75} \cdot 16 = \frac{16}{43.75};$\
$\mathbb{D}\eta = \frac{1}{43.75} \cdot 4 = \frac{4}{43.75};$\
$\mathop{\mathrm{cov}}\limits(\xi, \eta) = \mathop{\mathrm{cov}}\limits(\eta, \xi) = \frac{1}{43.75} \cdot 4.5 = \frac{4.5}{43.75};$\
${\mathop{\mathrm{\rho}}\limits(\xi, \eta)} = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}} = \frac{ \frac{4.5}{43.75}}{\sqrt{\frac{16}{43.75} \cdot \frac{4}{43.75}}} = 0.5625;$\
$C = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} = \frac{1}{2\pi \cdot \sqrt{\frac{1}{43.75}}} = \frac{5\sqrt{7}}{4\pi}.$\
Построим график плотности распределения $p_{\xi,\eta}:$

```python
#Импортируем все нужные библиотеки
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal
```

```python
mu = np.array([[-2], [-1]])
Sigma = np.array([[16 / 43.75, 4.5 / 43.75],
                  [4.5 / 43.75, 4 / 43.75]])

def p_xi_eta(x, y):
    return ((5 * np.sqrt(7)) / (4 * np.pi)) * np.exp(-0.5 * (4 * (x + 2)**2 - 9 * (x + 2) * (y + 1) + 16 * (y + 1)**2))

x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)

z_1 = p_xi_eta(X, Y)
z_2 = multivariate_normal.pdf(np.c_[X.ravel(), Y.ravel()], mean=mu.ravel(), cov=Sigma).reshape(X.shape)

fig = plt.figure(figsize=(25, 8))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, z_1, cmap='viridis', edgecolor='black')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Плотность распределения')
ax1.set_title('Плотность распределения, представленная графически с помощью функции плотности')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, z_2, cmap='viridis', edgecolor='black')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')


ax2.set_zlabel('Плотность распределения', labelpad=10)

ax2.set_title('Графическое представление многомерного распределения,\n построенное по ковариационной матрице и вектору математических ожиданий')


plt.show()
```

$p_{\vec{z}}(\vec{z}) = \frac{1}{2\pi \cdot 9} \cdot \exp\left( -\frac{1}{2} \cdot \left[4\xi^2 + 7\xi + 16\eta^2 + 14\eta - 9\xi\eta + 14\right] \right)$\
Построим график плотности полученного совместного распределения:

```python
def p_z(xi, eta):
    return (1 / (2 * np.pi * 9)) * np.exp(-0.5 * ((4 * xi**2) + (7 * xi) + (16 * eta**2) + (14 * eta) - (9 * xi * eta) + 14))

xi = np.linspace(-3, 3, 50)
eta = np.linspace(-3, 3, 50)

X, Y = np.meshgrid(xi, eta)
Z = p_z(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='white')

ax.set_xlabel('xi')
ax.set_ylabel('eta')
ax.set_zlabel('Плотность распределения')
ax.set_title('График плотности\nсовместного распределения p_z')

plt.show()
```


###5. Найдем условное распределение $\xi$ при условии $\eta$.
Для многомерного нормального распределения условное распределение также является нормальным. Условное распределение $\xi | \eta = y$ имеет форму:\
$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$, где\
$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{\sigma_{\xi\eta}}{\sigma_{\eta}^2} \cdot (y - \mu_{\eta}),$\
$\mathbb{D}[\xi|\eta] = \sigma_{\xi}^2 - \frac{\sigma_{\xi\eta}^2}{\sigma_{\eta}^2}.$\
Вычислим $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta).$\
$\mathbb{E}[\xi|\eta] = -2 + \frac{\frac{4.5}{43.75}}{\frac{4}{43.75}} \cdot (y_0 + 1) = -0.875 + 1.125y_0,$\
$\mathbb{D}[\xi|\eta] = \frac{16}{43.75} - \frac{(\frac{4.5}{43.75})^2}{\frac{4}{43.75}} = \frac{1}{4}$\
Таким образом, условное распределение $\xi | \eta = y_0$ имеет вид:\
$\xi | \eta = y \sim \mathcal{N}(-0.875 + 1.125y_0, \frac{1}{4})$,\
функция плотности будет иметь вид:\
$p_{\xi|\eta=y_0}(x) = \frac{1}{\sqrt{2\pi \sigma ^2 _{\xi | \eta}}} exp(-\frac{(x - \mu_{\xi | \eta = y_0})^2}{2\sigma ^2 _{\xi | \eta} });$\
$p_{\xi|\eta=y_0}(x) = \frac{1}{\sqrt{2\pi \sigma ^2 _{\xi | \eta}}} exp(-\frac{(x - (-0.875 + 1.125y_0))^2}{2 \cdot \frac{1}{4} }).$
