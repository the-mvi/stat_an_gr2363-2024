## Основные теоретические положения

Нормальное распределение двумерного случайного вектора определяется функцией плотности:
$p(\vec{x}) = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \exp\left( -\frac{1}{2} \cdot (\vec{x} - \vec{\mu})^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \vec{\mu}) \right), $
где: $\vec{x}=\begin{pmatrix} 
\xi \\ 
\eta 
\end{pmatrix},
\vec{\mu} = \mathbb{E}\begin{pmatrix} 
\xi \\ 
\eta 
\end{pmatrix},
\Sigma = \begin{pmatrix} 
\mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\
\mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta 
\end{pmatrix}.$

Аффинные преобразования:

$\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$ - случайный вектор;

$\vec{\xi}_{st} \sim\mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ — стандартный нормальный случайный вектор;

$\vec{\xi}_{st} = A \cdot\vec{\xi} + b$ — афинное преобразование случайного вектора;

$A\cdot\vec{\xi}+b \sim \mathcal{N}(A\cdot\vec{\mu}+b, A\cdot\Sigma \cdot A^{\intercal})$

где
$\vec{\mu}_{st}=A\cdot\vec{\mu}+b ; \Sigma_{st} = A\cdot\Sigma \cdot A^{\intercal}$

Ортогональные преобразования:

$Q^{\intercal}$ — матрица из собственных векторов матрицы $\Sigma$

$Q \cdot \Sigma \cdot Q^{\intercal} = D$ 

где: $D=\begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix}.$

$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} \cdot \exp\left(-\frac{1}{2} \cdot \left[ \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$ - плотность распределения

Cовместное распределение случайного вектора:

$\vec{b} = A \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$ — случайный вектор,

$\vec{b} \sim \mathcal{N}(A \cdot \vec{\mu}, A \cdot \Sigma \cdot A^{\intercal})$

Условное распределение:

$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$

$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{cov({\xi}{\eta})}{\mathbb{D}_{\eta}} \cdot (y - \mu_{\eta})$

$\mathbb{D}[\xi|\eta] = \mathbb{D}_{\xi} - \frac{cov({\xi}{\eta})^2}{\mathbb{D}_{\eta}}$

$\begin{cases}
\mathbb{E}(X) = \mathbb{E}(-3\xi-2\eta) = -3\mathbb{E}(\xi)-2\mathbb{E}(\eta) = -3 +2 = -1\\
\mathbb{E}(Y) = \mathbb{E}(-\xi-3\eta) = -\mathbb{E}(\xi)-3\mathbb{E}(\eta) = -1 +3 = 2\end{cases}$

$\begin{cases}
\mathbb{D}(X) = \mathbb{D}(-3\xi-2\eta) = 9\cdot\mathbb{D}(\xi) +4\cdot\mathbb{D}(\eta) = 9\cdot\frac{4}{25} +4\cdot\frac{28}{75} = \frac{44}{15}\\
\mathbb{D}(Y) = \mathbb{D}(-\xi-3\eta) = \mathbb{D}(\xi) + 9\cdot\mathbb{D}(\eta) = \frac{4}{25} +9\cdot\frac{28}{75} = \frac{88}{25}\end{cases}$

$cov(-3\xi-2\eta, -\xi -3\eta) = \mathbb{E}((-3\xi-2\eta)\cdot(-\xi -3\eta)) - \mathbb{E}(-3\xi-2\eta)\cdot\mathbb{E}(-\xi -3\eta) = \mathbb{E}(3\xi^2 +11\xi\eta +6\eta^2) - \mathbb{E}(-3\xi-2\eta)\cdot\mathbb{E}(-\xi -3\eta) = $

$= 3\mathbb{D}(\xi) +6\mathbb{D}(\eta) +11(\mathbb{E}(\xi\cdot\eta)-\mathbb{E}(\xi)\mathbb{E}(\eta))$; т.к. компоненты $\xi$ и $\eta$ независимые $\implies \mathbb{E}(\xi\cdot\eta)-\mathbb{E}(\xi)\mathbb{E}(\eta) = 0$

$cov(-3\xi-2\eta, -\xi -3\eta) = 3\cdot\frac{4}{25} +6\cdot\frac{28}{75} = \frac{68}{25}$

$\Sigma_{XY} = \frac{1}{5}\begin{pmatrix}
\frac{44}{3} & \frac{68}{5} \\
\frac{68}{5} & \frac{88}{5}\end{pmatrix}$

$P_{XY}(x,y) = \frac{1}{2\pi\sqrt{\mathbb{D}(x)\mathbb{D}(y)}\sqrt{1-r^2}}\cdot exp(-\frac{1}{2(1-r^2)}(\frac{(x-\mathbb{E}(x))^2}{\mathbb{D}(x)}-\frac{2r(x-\mathbb{E}(x))(y-\mathbb{E}(y))}{\sqrt{\mathbb{D}(x)\mathbb{D}(y)}}+\frac{(y-\mathbb{E}(y))^2}{\mathbb{D}(y)}))$

$r=\frac{cov(x,y)}{\sqrt{\mathbb{D}(x)\mathbb{D}(y)}} = \frac{17\sqrt{30}}{110}$

$p(x,y) = \frac{\frac{28\sqrt{21}}{75}}{\frac{10976\pi}{1875}}\cdot exp(-\frac{605}{343}\cdot\frac{5}{484}\cdot(33x^2 + 168x + 245 - 51xy - 161y + \frac{55}{2}y^2))$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def p_xy(x, y):
    c = (28 * np.sqrt(21) / 75) / (10976 * np.pi / 1875)
    exp = - (605 / 343) * (5 / 484) * (33 * x**2 + 168 * x + 245 - 51 * x * y - 161 * y + (55 / 2) * y**2)
    return c * np.exp(exp)

x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)

Z = p_xy(X, Y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='white')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('p(x, y)')

plt.title('Совместное распределение p(x, y)')
plt.show()
```
