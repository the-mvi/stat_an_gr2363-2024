## Основные теоретические положения



Плотность нормального случайного вектора: $p_{\vec{\xi}}(x_1,...,x_n)=\frac{1}{(2\Pi)^{\frac{n}{2}}\sqrt{det\Sigma}}\exp\left(-\frac{1}{2}(\vec{X}-\mu)^T\Sigma^{-1}(\vec{X}-\mu)\right)$

Математическое ожидание:

$\mu = \mathbb E \vec{\xi} = (\mathbb E \xi_1, ..., \mathbb E \xi_n)^{\intercal}$

Матрица ковариаций:

$\Sigma = \begin{pmatrix}
\mathbb{D}_\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta)\\
\mathop{\mathrm{cov}}\nolimits(\xi, \eta) & \mathbb{D}_\eta
\end{pmatrix}$

СКО:

$\sigma_{ij} = \mathop{cov}(\xi_i, \xi_j)$

Аффинное преобразование - преобразование вида:

$ \begin{pmatrix}
x\\
y
\end{pmatrix} = A \begin{pmatrix}
u\\
v
\end{pmatrix} + \begin{pmatrix}
\alpha\\
\beta
\end{pmatrix} $

Условное распределение $ \xi $ при условии $ \eta$:

$ \rho_{\xi}(\eta=y_0|x)=\frac{\rho_{\xi,\eta}(x, y_0)}{\rho_{\eta}(y_0)} $

Плотность двумерного нормального распределения имеет вид:    $\qquad \rho_{\xi,\eta}(x,y)=C\cdot \exp\left(-\frac{1}{2}\left( 3x^2-3xy+7y^2+12x+19y+37 \right)\right)$
1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $\rho_{\xi, \eta}$

   $ {q}(x,y)= 3x^2-3xy+7y^2+12x+19y+37=3(x+3)^{2}+2 \cdot (-\frac{3}{2})(y+2)(x+3)+7(y+2)^{2} $

   Вектор мат.ожидания:

   $ \mathbb{E} \begin{pmatrix}\xi \\ \eta \end{pmatrix}=\begin{pmatrix} -3 \\ -2 \end{pmatrix}; $

   Ковариационная матрица: 

   $ \Sigma^{-1}=\begin{pmatrix}3 & -1.5 \\ -1.5 & 7\end{pmatrix} \Rightarrow  \Sigma=\frac{1}{18,75}\begin{pmatrix}7 & 1.5 \\ 1.5 & 3\end{pmatrix}; $

   $ \mathbb{D}\xi=\frac{7}{18.75}, \quad \mathbb{D}\eta=\frac{3}{18.75}; $

   $ \mathop{\mathrm{cov}}\nolimits(\xi, \eta)=\frac{6}{73}; $

   $ \rho(\xi, \eta)=\frac{\sqrt{21}}{14}. $

   $ C=\frac{1}{2\pi\sqrt{\frac{1}{18,75}}} $

```python
import numpy as np
import matplotlib.pyplot as plt
import math

X, Y = np.meshgrid(np.linspace(-7, 7, 400), np.linspace(-7, 7, 400))
f = (1/(2*math.pi*math.sqrt(1/18.75)))*np.exp(-0.5*(3*X**2-3*X*Y+7*Y**2+12*X+19*Y+37))
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, f)
plt.xlabel('ось X')
plt.ylabel('ось Y')

plt.show()
```
