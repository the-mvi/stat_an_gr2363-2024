## Основные теоретические положения
Нормальное распределение двумерного случайного вектора определяется функцией плотности:
$p(\vec{x}) = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} \exp\left( -\frac{1}{2} \cdot (\vec{x} - \vec{\mu})^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \vec{\mu}) \right)$
где:\
$\vec{x} = \begin{pmatrix} \xi \\ \eta \end{pmatrix},$\
$\vec{\mu} = \mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix},$\
$\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}.$\
$\quad$\
Афинное преобразование:\
$\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$ — исходный случайный вектор;\
$\vec{\xi}_{st} \sim \mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ — стандартный нормальный случайный вектор;\
$\vec{\xi}_{st} = A \cdot \vec{\xi} + b$ — афинное преобразование случайного вектора;\
$\vec{\mu}_{st} = A \cdot \vec{\mu} + b$ — получение стандартного нормального вектора математических ожиданий из исходного;\
$\Sigma_{st} = A \cdot \Sigma \cdot A^{\intercal}$ — получение стандартной нормальной ковариационной матрицы из исходной.\
$\quad$\
Ортогональное преобразование, переводящее соответствующий центрированный вектор в вектор с независимыми компонентами:\
$Q^{\intercal}$ — матрица из собственных векторов матрицы $\Sigma;$\
При вычислении выражения $Q \cdot \Sigma \cdot Q^{\intercal}$ должна получится матрица $D$ вида: $\begin{pmatrix} \ldots & 0 \\ 0 & \ldots \end{pmatrix};$\
Матрица $D$ является диагональной матрицей из собственных значений $\Sigma^{-1},$\
при этом $\Sigma = Q \cdot D^{-1} \cdot Q^{\intercal},$\
Плотность распределения независимых компонентов центрированного вектора будет иметь вид:\
$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} \cdot \exp\left(-\frac{1}{2} \cdot \left[  \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$\
$\quad$\
Cовместное распределение случайного вектора:\
Пусть $\vec{z} = A \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$ — случайный вектор,\
тогда $\vec{z} \sim \mathcal{N}(A \cdot \vec{\mu}, A \cdot \Sigma \cdot A^{\intercal})$\
$\quad$\
Условное распределение для нормального:\
Для многомерного нормального распределения условное распределение также является нормальным.\
Условное распределение $\xi | \eta = y$ имеет следующую форму:\
$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$, где\
$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{\sigma_{\xi\eta}}{\sigma_{\eta}^2} \cdot (y - \mu_{\eta}),$\
$\mathbb{D}[\xi|\eta] = \sigma_{\xi}^2 - \frac{\sigma_{\xi\eta}^2}{\sigma_{\eta}^2}.$

###1. Вычислим вектор математических ождианий и ковариационные характеристики данного случайного вектора, для этого сначала приведем квадратичную форму $Q(x, y)$ к каноническому виду:
$Q(x, y) = 4x^2 + 4xy + 7y^2 - 12x + 6y + 15 =$\
$= 4(x^2 - 4x + 4) + 4xy + 7y^2 + 4x + 6y - 1 =$\
$= 4(x^2 - 4x + 4) + 7(y^2 + 2y + 1) + 4xy + x - 8y - 8 =$\
$= 4(x^2 - 4x + 4) + 7(y^2 + 2y + 1) + 4(xy + x - 2y - 2) =$\
$= 4(x^2 - 4x + 4) + 7(y^2 + 2y + 1) + 4\left(x(y + 1) - 2(y + 1)\right) =$\
$= 4(x - 2)^2 + 4(x - 2)(y + 1) + 7(y + 1)^2$\
В результате приведения к каноническому виду квадратичной формы получаем:\
$p_{\xi,\eta}(x, y) = C \cdot \exp\left(-\frac{1}{2} \cdot \left[4(x - 2)^2 + 4(x - 2)(y + 1) + 7(y + 1)^2\right]\right).$\
Следовательно, вектор математических ожиданий данного случайного вектора имеет вид:\
$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 2 \\ -1 \end{pmatrix}.$\
Также получим матрицу $\Sigma^{-1}:$\
$\Sigma^{-1} = \begin{pmatrix} 4 & 2 \\ 2 & 7 \end{pmatrix}.$\
Найдем ковариационную матрицу $\Sigma:$\
$\Sigma = \frac{1}{|\Sigma^{-1}|} \cdot \begin{pmatrix} 7 & -2 \\ -2 & 4 \end{pmatrix} = \frac{1}{24} \cdot \begin{pmatrix} 7 & -2 \\ -2 & 4 \end{pmatrix}.$\
Зная, что ковариционная матрица имеет вид: $\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix},$ запишем ковариационные характеристики:\
$\mathbb{D}\xi = \frac{1}{24} \cdot 7 = \frac{7}{24};$\
$\mathbb{D}\eta = \frac{1}{24} \cdot 4 = \frac{4}{24} = \frac{1}{6};$\
$\mathop{\mathrm{cov}}\limits(\xi, \eta) = \mathop{\mathrm{cov}}\limits(\eta, \xi) = \frac{1}{24} \cdot -2 = -\frac{2}{24} = -\frac{1}{12};$\
${\mathop{\mathrm{\rho}}\limits(\xi, \eta)} = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}} = \frac{-\frac{1}{12}}{\sqrt{\frac{7}{24} \cdot \frac{1}{6}}} = -\frac{\sqrt{7}}{7};$\
$C = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} = \frac{1}{2\pi \cdot \sqrt{\frac{1}{24}}} = \frac{\sqrt{6}}{\pi}.$\
Построим график плотности распределения $p_{\xi,\eta}:$

```r
#Установка пакетов
#install.packages("plot3D")
#install.packages("mvtnorm")

library(plot3D)
library(mvtnorm)

mu <- matrix(c(2, -1), ncol = 1)
Sigma <- matrix(c(7/24, -1/12, -1/12, 1/6), nrow = 2)

p_xi_eta <- function(x, y){(sqrt(6) / pi) * exp(-0.5 * (4 * (x - 2)^2 + 4 * (x - 2) * (y + 1) + 7 * (y + 1)^2))}

x <- seq(-3, 3, length.out = 50)
y <- seq(-3, 3, length.out = 50)

z_1 <- outer(x, y, p_xi_eta)

grid <- expand.grid(x = x, y = y)
z_2 <- matrix(dmvnorm(grid, mean = mu, sigma = Sigma), nrow = length(x))


#График строится напрямую по функции плотности p_xi_eta
persp3D(x, y, z_1, col = "azure", border = "black", shade = 0.0,
theta = 30, phi = 20, ltheta = 90, lphi = 180,
xlab = "X", ylab = "Y", zlab = "Плотность распределения",
main = "График плотности распределения\nна основе функции плотности")

#График строится с помощью пакета mvtnorm (Multivariate Normal and t Distributions), используя ковариационную матрицу Sigma и ветор математических ожиданий mu
persp3D(x, y, z_2, col = "azure", border = "black", shade = 0.0,
theta = 30, phi = 20, ltheta = 90, lphi = 180,
xlab = "X", ylab = "Y", zlab = "Плотность распределения",
main = "График плотности распределения\nна основе ковариационной матрицы\nи вектора математических ожиданий")
```

Афинное преобразование осуществляется по следующим формулам:\

$\vec{\xi}_{st} \sim \mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ — стандартный нормальный случайный вектор;\
$\vec{\xi}_{st} = A \cdot \vec{\xi} + b$ — афинное преобразование случайного вектора;\
$\vec{\mu}_{st} = A \cdot \vec{\mu} + b$ — получение стандартного нормального вектора математических ожиданий из исходного;\

###3. Найдем ортогональное преобразование, переводящее соответствующий центрированный вектор в вектор с независимыми компонентами.
Найдем собственные числа и собственные вектора для матрицы $\Sigma^{-1},$ для этого решим уравнение:\
$\mathop{\mathrm{det}}(\Sigma^{-1} - \lambda \cdot E) = 0$\
$\left|\begin{pmatrix} 4 & 2 \\ 2 & 7 \end{pmatrix} - \begin{pmatrix} \lambda & 0 \\ 0 & \lambda \end{pmatrix}\right| = \left| \begin{pmatrix} 4 - \lambda & 2 \\ 2 & 7 - \lambda\end{pmatrix} \right| = 0$\

Вычислим $\lambda_{1}$ и $\lambda_{2}:$

Решим системы уравнений для $\lambda_{1}$ и $\lambda_{2}:$\

 \begin{cases}

$\vec{v_1} = \begin{pmatrix} 1 \\ 2 \end{pmatrix} \cdot \frac{1}{\sqrt{5}}$\
2) Для $\lambda_{2}:$\
$\begin{equation*}
 \begin{cases}
   u_2 + 2 \cdot v_2 = 0
   \\
   2 \cdot u_2 + 4 \cdot v_2 = 0
 \end{cases}
\end{equation*}$\
$\begin{equation*}
 \begin{cases}

   \\

 \end{cases}
\end{equation*}$\
$\begin{equation*}
 \begin{cases}
   \sqsupset \beta \in \mathbb{R}
   \\

 \end{cases}
\end{equation*}$\
$\vec{v_2} = \begin{pmatrix} -2 \\ 1 \end{pmatrix} \cdot \beta, \beta \in \mathbb{R}$\

$\vec{v_2} = \begin{pmatrix} -2 \\ 1 \end{pmatrix} \cdot \frac{1}{\sqrt{5}}$\
Составим матрицу $Q^{\intercal}$ из собственных векторов:\
$Q^{\intercal} = \begin{pmatrix} 1 & -2 \\ 2 & 1 \end{pmatrix} \cdot \frac{1}{\sqrt{5}},$ тогда матрица $Q$ будет иметь вид:\
$Q = \begin{pmatrix} 1 & 2 \\ -2 & 1 \end{pmatrix} \cdot \frac{1}{\sqrt{5}}.$\
Произведем проверку:\
При вычислении выражения $Q \cdot \Sigma \cdot Q^{\intercal}$ должна получится матрица вида: $\begin{pmatrix} \ldots & 0 \\ 0 & \ldots \end{pmatrix};$\
$\frac{1}{\sqrt{5} \cdot 24 \cdot \sqrt{5}} \cdot \begin{pmatrix} 1 & 2 \\ -2 & 1 \end{pmatrix} \cdot \begin{pmatrix} 7 & -2 \\ -2 & 4 \end{pmatrix} \cdot \begin{pmatrix} 1 & -2 \\ 2 & 1 \end{pmatrix} = \begin{pmatrix} \frac{1}{3} & 0 \\ 0 & \frac{1}{8} \end{pmatrix}$\
Результат удовлетворяет условию, значит преобразование выполнено верно.\
Так как полученная матрица $D = \begin{pmatrix} \frac{1}{3} & 0 \\ 0 & \frac{1}{8} \end{pmatrix}$ является диагональной матрицей из собственных значений $\Sigma^{-1},$\
$\Sigma = Q \cdot D^{-1} \cdot Q^{\intercal},$\
где: $D^{-1} = \frac{1}{|D|} \cdot \begin{pmatrix} \frac{1}{3} & 0 \\ 0 & \frac{1}{8} \end{pmatrix} = 24 \cdot \begin{pmatrix} \frac{1}{3} & 0 \\ 0 & \frac{1}{8} \end{pmatrix} = \begin{pmatrix} 3 & 0 \\ 0 & 8 \end{pmatrix}.$\
Таким образом, плотность распределения независимых компонентов центрированного вектора будет иметь вид:\
$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} \cdot \exp\left(-\frac{1}{2} \cdot \left[  \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right) = \frac{\sqrt{6}}{\pi} \cdot\exp\left(-\frac{1}{2} \cdot \left[ 3{x'}^2 + 8{y'}^2 \right]\right).$\
Построим график плотности полученного распределения:

```r
#Установка пакетов
#install.packages("plot3D")

library(plot3D)

p_xi_eta <- function(x, y){(sqrt(6) / pi) * exp(-0.5 * (3 * x^2 + 8 * y^2))}

x <- seq(-3, 3, length.out = 50)
y <- seq(-3, 3, length.out = 50)

z <- outer(x, y, p_xi_eta)

persp3D(x, y, z, col = "azure", border = "black", shade = 0.0,
theta = 30, phi = 20, ltheta = 90, lphi = 180,
xlab = "X", ylab = "Y", zlab = "Плотность распределения",
main = "График плотности распределения\nнезависимых компонентов центрированного вектора")

persp3D(x, y, z, col = "azure", border = "black", shade = 0.0,
theta = 90, phi = 0, ltheta = 90, lphi = 180,
xlab = "X", ylab = "Y", zlab = "Плотность распределения",
main = "График плотности распределения\nнезависимых компонентов центрированного вектора")

persp3D(x, y, z, col = "azure", border = "black", shade = 0.0,
theta = 180, phi = 0, ltheta = 90, lphi = 180,
xlab = "X", ylab = "Y", zlab = "Плотность распределения",
main = "График плотности распределения\nнезависимых компонентов центрированного вектора")
```

###4. Вычислим характеристики совместного распределения случайного вектора $(-\xi + 2\eta, -2\xi - \eta)^{\intercal}$ и запишем его плотность.

Пусть $\vec{z}=\begin{pmatrix} -\xi + 2\eta \\ -2\xi - \eta \end{pmatrix} = A \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix},$\
тогда $\vec{z} \sim \mathcal{N}(A \cdot \vec{\mu}, A \cdot \Sigma \cdot A^{\intercal})$, где матрица $A = \begin{pmatrix} -1 & 2 \\ -2 & -1 \end{pmatrix}$\
Обозначим $A \cdot \vec{\mu} = \mathbb{E}[\vec{z}] = \vec{\mu}_{\vec{z}}$ и $A \cdot \Sigma \cdot A^{\intercal} = \Sigma_{\vec{z}}$\
Произведем расчеты:

```r
mu <- matrix(c(2, -1), ncol = 1)
Sigma <- matrix(c(7/24, -1/12, -1/12, 1/6), nrow = 2)
A <- matrix(c(-1, -2, 2, -1), nrow = 2)

mu_z <- A %*% mu
Sigma_z <- A %*% Sigma %*% t(A)

det_Sigma_z <- (Sigma_z[1] * Sigma_z[4] - (Sigma_z[2] * Sigma_z[3]))

cat("mu_z:","\n")
prmatrix(mu_z)

cat("Sigma_z ⋅ 24:","\n")
prmatrix(Sigma_z * 24)

cat("det_Sigma_z ⋅ 24:", det_Sigma_z * 24)
```

Исходя из расчетов:\
$\vec{\mu}_{\vec{z}} = \begin{pmatrix} -1 & 2 \\ -2 & -1 \end{pmatrix} \cdot \begin{pmatrix} 2 \\ -1 \end{pmatrix} = \begin{pmatrix} -4 \\ -3 \end{pmatrix},$\
$\Sigma_{\vec{z}} = \frac{1}{24} \cdot \begin{pmatrix} -1 & 2 \\ -2 & -1 \end{pmatrix} \cdot \begin{pmatrix} 7 & -2 \\ -2 & 4 \end{pmatrix} \cdot \begin{pmatrix} -1 & -2 \\ 2 & -1 \end{pmatrix} = \frac{1}{24} \cdot \begin{pmatrix} 31 & 12 \\ 12 & 24 \end{pmatrix}.$\
Имея ковариационную матрицу, можем выписать харктеристики совместного распределения:\
$\mathbb{D}[-\xi + 2\eta] = \frac{1}{24} \cdot 31 = \frac{31}{24};$\
$\mathbb{D}[-2\xi - \eta] = \frac{1}{24} \cdot 24 = \frac{24}{24} = 1;$\
$\mathop{\mathrm{cov}}\limits(-\xi + 2\eta, -2\xi - \eta) = \frac{1}{24} \cdot 12 = \frac{12}{24} = \frac{1}{2};$\
${\mathop{\mathrm{\rho}}\limits(\xi, \eta)} = \frac{\mathop{\mathrm{cov}}\limits(-\xi + 2\eta, -2\xi - \eta)}{\sqrt{\mathbb{D}[-\xi + 2\eta] \cdot \mathbb{D}[-2\xi - \eta]}} = \frac{\frac{1}{2}}{\sqrt{\frac{31}{24} \cdot 1}} = \frac{\sqrt{186}}{31};$\
Теперь можем найти функцию плотности:\
$p_{\vec{z}}(\vec{z}) = \frac{1}{2\pi \cdot \sqrt{|\Sigma_{\vec{z}}|}} \cdot \exp\left( -\frac{1}{2} \cdot (\vec{z} - \vec{\mu}_{\vec{z}})^{\intercal} \cdot \Sigma_{\vec{z}}^{-1} \cdot (\vec{z} - \vec{\mu}_{\vec{z}}) \right);$\
$\Sigma_{\vec{z}}^{-1} = \frac{1}{|\Sigma_{\vec{z}}|} \cdot \begin{pmatrix} 24 & -12 \\ -12 & 31 \end{pmatrix} = \frac{24}{25} \cdot \begin{pmatrix} 24 & -12 \\ -12 & 31 \end{pmatrix};$\
$p_{\vec{z}}(\vec{z}) = \frac{1}{2\pi \cdot \sqrt{\frac{25}{24}}} \cdot \exp\left( -\frac{1}{2} \cdot \frac{24}{25} \cdot \left[ \begin{pmatrix} -\xi + 2\eta \\ -2\xi - \eta \end{pmatrix} - \begin{pmatrix} -4 \\ -3 \end{pmatrix} \right]^{\intercal} \cdot \begin{pmatrix} 24 & -12 \\ -12 & 31 \end{pmatrix} \cdot \left[ \begin{pmatrix} -\xi + 2\eta \\ -2\xi - \eta \end{pmatrix} - \begin{pmatrix} -4 \\ -3 \end{pmatrix} \right] \right)=$\
$= \frac{1}{2\pi \cdot \sqrt{\frac{25}{24}}} \cdot \exp\left( -\frac{1}{2} \cdot \frac{24}{25} \cdot \begin{pmatrix} -\xi + 2\eta + 4 & -2\xi - \eta + 3 \end{pmatrix} \cdot \begin{pmatrix} 24 & -12 \\ -12 & 31 \end{pmatrix} \cdot \begin{pmatrix} -\xi + 2\eta + 4 \\ -2\xi - \eta + 3 \end{pmatrix} \right)=$\
$= \frac{1}{2\pi \cdot \sqrt{\frac{25}{24}}} \cdot \exp\left( -\frac{1}{2} \cdot \frac{24}{25} \cdot \begin{pmatrix} 60\eta + 60 & -50\xi - 55\eta + 45 \end{pmatrix} \cdot \begin{pmatrix} -\xi + 2\eta + 4 \\ -2\xi - \eta + 3 \end{pmatrix} \right)=$
$= \frac{1}{2\pi \cdot \sqrt{\frac{25}{24}}} \cdot \exp\left( -\frac{1}{2} \cdot \frac{24}{25} \cdot \left[100\xi^2 - 300\xi + 175\eta^2 + 150\eta + 100\xi\eta + 375\right] \right);$
$p_{\vec{z}}(\vec{z}) = \frac{1}{2\pi \cdot \sqrt{\frac{25}{24}}} \cdot \exp\left( -\frac{1}{2} \cdot \left[96\xi^2 - 288\xi + 168\eta^2 + 144\eta + 96\xi\eta + 360\right] \right)$\
Построим график плотности полученного совместного распределения:

```r
#Установка пакетов
#install.packages("plot3D")

library(plot3D)

p_z <- function(xi, eta){(1 / (2 * pi * sqrt(25/24))) * exp(-0.5 * ((96 * xi^2) - (288 * xi) + (168 * eta^2) + (144 * eta) + (96 * xi * eta) + 360))}

xi <- seq(-3, 3, length.out = 50)
eta <- seq(-3, 3, length.out = 50)

z <- outer(xi, eta, p_z)

persp3D(xi, eta, z, col = "azure", border = "black", shade = 0.0,
theta = 30, phi = 20, ltheta = 90, lphi = 180,
xlab = "xi", ylab = "eta", zlab = "Плотность распределения",
main = "График плотности\nсовместного распределения p_z")
```


###5. Найдем условное распределение $\xi$ при условии $\eta$.
Для многомерного нормального распределения условное распределение также является нормальным.\
Условное распределение $\xi | \eta = y$ имеет следующую форму:\
$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$, где\
$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{\sigma_{\xi\eta}}{\sigma_{\eta}^2} \cdot (y - \mu_{\eta}),$\
$\mathbb{D}[\xi|\eta] = \sigma_{\xi}^2 - \frac{\sigma_{\xi\eta}^2}{\sigma_{\eta}^2}.$\
Вычислим $\mathbb{E}[\xi|\eta]$ и $\mathbb{D}[\xi|\eta]:$\
$\mathbb{E}[\xi|\eta] = 2 + \frac{-\frac{2}{24}}{\frac{4}{24}} \cdot (y + 1) = 2 -\frac{1}{2} \cdot (y + 1) = 2 -\frac{1}{2} \cdot y -\frac{1}{2} = \frac{3}{2} -\frac{1}{2} \cdot y;$\
$\mathbb{D}[\xi|\eta] = \frac{7}{24} - \frac{(-\frac{2}{24})^2}{\frac{4}{24}} = \frac{7}{24} - \frac{\frac{4}{576}}{\frac{4}{24}} = \frac{7}{24} - \frac{1}{24} = \frac{6}{24} = \frac{1}{4}.$\
Таким образом, условное распределение $\xi | \eta = y$ имеет вид:\
$\xi | \eta = y \sim \mathcal{N}(\frac{3}{2} -\frac{1}{2} \cdot y, \frac{1}{4})$
