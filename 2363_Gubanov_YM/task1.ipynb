## Основные теоретические положения

Нормальное распределение двумерного случайного вектора определяется функцией плотности:
$p(\vec{x}) = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \exp\left( -\frac{1}{2} \cdot (\vec{x} - \vec{\mu})^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \vec{\mu}) \right), $
где: $\vec{x}=\begin{pmatrix} 
\xi \\ 
\eta 
\end{pmatrix},
\vec{\mu} = \mathbb{E}\begin{pmatrix} 
\xi \\ 
\eta 
\end{pmatrix},
\Sigma = \begin{pmatrix} 
\mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\
\mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta 
\end{pmatrix}.$

Аффинные преобразования:

$\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$ - случайный вектор;

$\vec{\xi}_{st} \sim\mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ — стандартный нормальный случайный вектор;

$\vec{\xi}_{st} = A \cdot\vec{\xi} + b$ — афинное преобразование случайного вектора;

$A\cdot\vec{\xi}+b \sim \mathcal{N}(A\cdot\vec{\mu}+b, A\cdot\Sigma \cdot A^{\intercal})$

где
$\vec{\mu}_{st}=A\cdot\vec{\mu}+b ; \Sigma_{st} = A\cdot\Sigma \cdot A^{\intercal}$

Ортогональные преобразования:

$Q^{\intercal}$ — матрица из собственных векторов матрицы $\Sigma$

$Q \cdot \Sigma \cdot Q^{\intercal} = D$ 

где: $D=\begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix}.$

$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} \cdot \exp\left(-\frac{1}{2} \cdot \left[ \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$ - плотность распределения

Cовместное распределение случайного вектора:

$\vec{b} = A \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$ — случайный вектор,

$\vec{b} \sim \mathcal{N}(A \cdot \vec{\mu}, A \cdot \Sigma \cdot A^{\intercal})$

Условное распределение:

$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$

$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{cov({\xi}{\eta})}{\mathbb{D}_{\eta}} \cdot (y - \mu_{\eta})$

$\mathbb{D}[\xi|\eta] = \mathbb{D}_{\xi} - \frac{cov({\xi}{\eta})^2}{\mathbb{D}_{\eta}}$

Дано совместное распределение случайного вектора: $(-3\xi-2\eta, -\xi -3\eta)^T$

Пусть $\vec{\gamma}=\begin{pmatrix} -3\xi - 2\eta \\ -\xi -3\eta \end{pmatrix} = A \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix} => A = \begin{pmatrix} -3 & -2 \\ -1 & -3 \end{pmatrix}$

$\vec{\mu}_{\gamma} = \begin{pmatrix} -3 & -2 \\ -1 & -3 \end{pmatrix} \cdot \begin{pmatrix} 1 \\ -1 \end{pmatrix} = \begin{pmatrix} -1 \\ 2 \end{pmatrix}$

$\Sigma_{\gamma} = \frac{4}{75}\cdot\begin{pmatrix} -3 & -2 \\ -1 & -3 \end{pmatrix} \cdot \begin{pmatrix}7 & -\frac{3}{2}  \\-\frac{3}{2}  & 3\end{pmatrix}\cdot \begin{pmatrix} -3 & -1 \\ -2 & -3 \end{pmatrix} = \frac{4}{75} \cdot \begin{pmatrix} 57 & \frac{45}{2} \\ \frac{45}{2} & 25 \end{pmatrix}$

$\mathbb{D}[-3\xi  -2\eta] = \frac{4}{75} \cdot 57 = \frac{76}{25}$

$\mathbb{D}[-\xi - 3\eta] = \frac{4}{75} \cdot 25 = \frac{4}{3}$

$cov(-3\xi - 2\eta, -\xi - 3\eta) = \frac{4}{75} \cdot \frac{45}{2} = \frac{6}{5}$

$\rho(-\xi + 2\eta, -2\xi - \eta) = \frac{cov(-\xi + 2\eta, -2\xi - \eta)}{\sqrt{\mathbb{D}[-\xi + 2\eta] \cdot \mathbb{D}[-2\xi - \eta]}} = \frac{\frac{6}{5}}{\sqrt{\frac{76}{25} \cdot \frac{4}{3}}} = \frac{3\sqrt{57}}{38}$

$\Sigma_{\vec{z}}^{-1} = \frac{1}{|\Sigma_{\vec{z}}|} \cdot \begin{pmatrix} 25 & -\frac{45}{2} \\ -\frac{45}{2} & 57 \end{pmatrix} = \frac{75}{4} \cdot \begin{pmatrix} 25 & -\frac{45}{2} \\ -\frac{45}{2} & 57 \end{pmatrix}$

$p_{\vec{\gamma}}(\vec{\gamma}) = \frac{1}{2\pi \cdot \sqrt{| \Sigma_{\vec{\gamma}} |}} \cdot \exp\left( -\frac{1}{2} \cdot (\vec{\gamma} - \vec{\mu}_{\vec{\gamma}})^{T} \cdot \Sigma_{\vec{\gamma}}^{-1} \cdot (\vec{\gamma} - \vec{\mu}_{\vec{\gamma}}) \right)$

$p_{\vec{\gamma}}(\vec{\gamma}) = \frac{1}{2\pi \cdot \sqrt{\frac{4}{75}}} \cdot \exp\left( -\frac{1}{2} \cdot \frac{75}{4}\cdot \left[\begin{pmatrix} -3x - 2y \\ -x -3y \end{pmatrix} - \begin{pmatrix} -1 \\ 2 \end{pmatrix}\right]^{T} \cdot \begin{pmatrix} 25 & -\frac{45}{2} \\ -\frac{45}{2} & 57 \end{pmatrix} \cdot \left[\begin{pmatrix} -3x - 2y \\ -x -3y \end{pmatrix} - \begin{pmatrix} -1 \\ 2 \end{pmatrix}\right] \right)$

$p_{\vec{\gamma}}(\vec{\gamma}) = \frac{5\sqrt{3}}{4\pi} \cdot \exp\left( -\frac{1}{2} \cdot \frac{75}{4} \cdot \left[147x^2 +147xy -147x +343y^2 +539y +343\right] \right)$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def p_xy(x, y):
    c = (5 * np.sqrt(3)) / (4 * np.pi)
    exp = - (1 / 2) * (75 / 4) * (147 * x**2 + 147 * x * y - 147 * x + 343 * y**2 + 539 * y + 343)
    return c * np.exp(exp)

x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)

Z = p_xy(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='white')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('p(x, y)')

plt.title('Совместное распределение p(x, y)')
plt.show()
```
