Плотность двумерного нормального распределения имеет вид:

1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения$ \rho_{\xi,\eta} $.
   $ q(x, y)=2x^2-4xy+5y^2-4x-2y+5=2((x-y)-1)^2+3(y-1)^2=2(x-2)^2+2\cdot(-2)(x-2)(y-1)+5(y-1)^2 $

   Мат. ожидание:  $ \mathbb{E} \begin{pmatrix}\xi \\ \eta \end{pmatrix}=\begin{pmatrix} 2 \\ 1 \end{pmatrix}; $
  
   Обратная матрица ковариаций: $ \Sigma^{-1}=\begin{pmatrix}2 & -2 \\ -2 & 5\end{pmatrix} $
  
   Матрица ковариаций: $ \Sigma=\frac{1}{6}\begin{pmatrix}5 & 2 \\ 2 & 2\end{pmatrix}; $
  
   Дисперсия: $ \mathbb{D}\eta=\frac{1}{3}, \quad \mathbb{D}\xi=\frac{1}{3}; $
  
   Ковариация: $ \mathop{\mathrm{cov}}\nolimits(\xi, \eta)=\frac{1}{3}; $
  
   Коэффицент корреляции: $ \rho(\xi, \eta)=\frac{\sqrt{2}}{5}. $

   Константа: $ C=\frac{1}{2\pi\sqrt{\frac{1}{6}}} $

```python
!pip install matplotlib
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)
X, Y = np.meshgrid(x, y)
Z = (1/(2*np.pi*np.sqrt(1/6)))*np.exp(-0.5*(2*X**2-4*X*Y+5*Y**2-4*X-2*Y+5))
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.title('График плотности двумерного нормального распределения')
plt.show()
```

2. Найти аффиное преобразование, переводящее исходный случайный вектор в стандартный нормальный.
   $ q(x,y)=2((x-y)-1)^2+3(y-1)^2=(\sqrt{2}(x-y)-\sqrt{2})^2+(\sqrt{3}y-\sqrt{3})^2 $
   
   Аффиное преобразование: $ \begin{pmatrix}\xi_{st}  \eta_{st} \end{pmatrix}=\begin{pmatrix} \sqrt{2} & -\sqrt{2} \\ 0 & \sqrt{3}\end{pmatrix} \cdot \begin{pmatrix}\xi \\ \eta \end{pmatrix}+\begin{pmatrix}-\sqrt{2} \\ -\sqrt{3} \end{pmatrix}. $
   
   Проверка:
   
   $ A\vec{\xi}+b\sim{N}\left(A\mathbb{E}+b,A \mathbb{E} A^{\intercal}\right); $

   $ \mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix}=\begin{pmatrix} \sqrt{2} & -\sqrt{2} \\ 0 & \sqrt{3} \end{pmatrix} \cdot \begin{pmatrix} 2 \\ 1 \end{pmatrix} + \begin{pmatrix} -\sqrt{2} \\ -\sqrt{3} \end{pmatrix}=\begin{pmatrix} \sqrt{2} \\ \sqrt{3}\end{pmatrix} + \begin{pmatrix} -\sqrt{2} \\ -\sqrt{3} \end{pmatrix}=\begin{pmatrix} 0 \\ 0 \end{pmatrix} $

   $ \Sigma_{st}=\frac{1}{6}\begin{pmatrix} \sqrt{2} & -\sqrt{2} \\ 0 & \sqrt{3} \end{pmatrix} \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix} \begin{pmatrix} \sqrt{2} & 0 \\ -\sqrt{2} & \sqrt{3} \end{pmatrix}=\frac{1}{6} \begin{pmatrix} 3\sqrt{2} & 0 \\ 2\sqrt{3} & 2\sqrt{3} \end{pmatrix} \begin{pmatrix} \sqrt{2} & 0 \\ -\sqrt{2} & \sqrt{3} \end{pmatrix}= \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}. $

   Получен нулевой вектор и  единичная матрица, значит проверка выполнена успешно.
