1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$ .
$q(x,y)=7x^2-8xy+13y^2-6x-18y+12=7(x-1)^2+2(-4)(x-1)(y-1)+13(y-1)^2$

    Вектор математических ожиданий: $\mathbb{E} \begin{pmatrix}\xi \\ \eta \end{pmatrix}=\begin{pmatrix} 1 \\ 1 \end{pmatrix};$
    
    Обратная матрица ковариации: $\Sigma^{-1}=\begin{pmatrix}7 & -4 \\ -4 & 13\end{pmatrix};$
    
    Матрица ковариаций: $\Sigma = \frac{1}{75}\begin{pmatrix}13 & 4\\ 4 & 7\end{pmatrix};$
    
    Дисперсии: $\mathbb{D}\xi=\frac{13}{75}; \mathbb{D}\eta=\frac{7}{75};$
    
    Ковариация: $Cov(\xi,\eta) = \frac{4}{75};$
    
    Коэффициент корреляции: $\rho(\xi, \eta)=\frac{4\sqrt{91}}{91};$

    Константа: $ C=\frac{1}{2\pi\sqrt{\frac{1}{75}}}.$
    
    Построим график плотности соответственно:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-15, 15, 50)
y = np.linspace(-15, 15, 50)
X, Y = np.meshgrid(x, y)
Z = (1/(2*np.pi*np.sqrt(1/75)))*np.exp(-0.5*(7*X**2-8*X*Y+13*Y**2-6*X-2*18+12))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.title('График плотности двумерного нормального распределения')
plt.show()
```
