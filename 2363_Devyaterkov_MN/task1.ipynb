## Основные теоретические положения

Плотность нормального случайного вектора: $p_{\vec{\xi}}(x_1,...,x_n)=\frac{1}{(2\Pi)^{\frac{n}{2}}\sqrt{det\Sigma}}\exp\left(-\frac{1}{2}(\vec{X}-\mu)^T\Sigma^{-1}(\vec{X}-\mu)\right)$

Математическое ожидание:

$\mu = \mathbb E \vec{\xi} = (\mathbb E \xi_1, ..., \mathbb E \xi_n)^{\intercal}$

Матрица ковариаций:

$\Sigma = \begin{pmatrix}
\mathbb{D}_\xi & \mathop{\mathrm{cov}}\nolimits(\xi, \eta)\\
\mathop{\mathrm{cov}}\nolimits(\xi, \eta) & \mathbb{D}_\eta
\end{pmatrix}$

СКО:

$\sigma_{ij} = \mathop{cov}(\xi_i, \xi_j)$

Аффинное преобразование - преобразование вида:

$ \begin{pmatrix}
x\\
y
\end{pmatrix} = A \begin{pmatrix}
u\\
v
\end{pmatrix} + \begin{pmatrix}
\alpha\\
\beta
\end{pmatrix} $

Условное распределение $ \xi $ при условии $ \eta$:

$ \rho_{\xi}(\eta=y_0|x)=\frac{\rho_{\xi,\eta}(x, y_0)}{\rho_{\eta}(y_0)} $

$cov(2\xi-2\eta, -5\xi -3\eta) = \mathbb{E}((2\xi-2\eta)\cdot(-5\xi -3\eta)) - \mathbb{E}(2\xi-2\eta)\cdot\mathbb{E}(-5\xi -3\eta) = \mathbb{E}(-10\xi^2 +4\xi\eta +6\eta^2) - \mathbb{E}(2\xi-2\eta)\cdot\mathbb{E}(-5\xi -3\eta) = $

$P_{XY}(x,y) = \frac{1}{2\pi\sqrt{\mathbb{D}(x)\mathbb{D}(y)}\sqrt{1-r^2}}\cdot exp(-\frac{1}{2(1-r^2)}(\frac{(x-\mathbb{E}(x))^2}{\mathbb{D}(x)}-\frac{2r(x-\mathbb{E}(x))(y-\mathbb{E}(y))}{\sqrt{\mathbb{D}(x)\mathbb{D}(y)}}+\frac{(y-\mathbb{E}(y))^2}{\mathbb{D}(y)}))$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def p(x, y):
    const = (4 * np.sqrt(6559)) / 75
    coeff = (209888 * np.pi) / 5626
    exponent = -((7575*x**2+68325*x+359850-2925*x*y-44850*y+1500*y**2)/26236)
    return (const / coeff) * np.exp(exponent)

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

Z = p(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Плотность')

fig.colorbar(surf, shrink=0.3, aspect=5)

plt.show()
```

Условное распределение $\xi | \eta = y_{0}$ имеет следующую форму: $\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$

$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{cov({\xi}{\eta})}{\mathbb{D}_{\eta}} \cdot (y - \mu_{\eta})$

$\mathbb{D}[\xi|\eta] = \mathbb{D}_{\xi} - \frac{cov({\xi}{\eta})^2}{\mathbb{D}_{\eta}}$

$\mathbb{E}[\xi|\eta] = -2 + \frac{\frac{2}{25}}{\frac{4}{25}} \cdot (y_{0} + 1) = -2 +\frac{1}{2} \cdot (y_{0} + 1) = 2 +\frac{1}{2} \cdot y_{0} -\frac{1}{2} = \frac{3}{2} +\frac{1}{2} \cdot y_{0}$

$\mathbb{D}[\xi|\eta] = \frac{28}{75} - \frac{(\frac{2}{25})^2}{\frac{4}{25}} = \frac{1}{3}$

$\xi | \eta = y_{0} \sim \mathcal{N}(\frac{3}{2} +\frac{1}{2} \cdot y_{0}, \frac{1}{3})$
