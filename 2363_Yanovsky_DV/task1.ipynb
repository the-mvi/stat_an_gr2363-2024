## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–≤—É–º–µ—Ä–Ω–æ–≥–æ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏:
$p(\vec{x}) = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \exp\left( -\frac{1}{2} \cdot (\vec{x} - \vec{\mu})^{\intercal} \cdot \Sigma^{-1} \cdot (\vec{x} - \vec{\mu}) \right)$
–≥–¥–µ:\
$\vec{x} = \begin{pmatrix} \xi \\ \eta \end{pmatrix},$\
$\vec{\mu} = \mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix},$\
$\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix}.$\
$\quad$\
–ê—Ñ–∏–Ω–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ:\
$\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$ ‚Äî –∏—Å—Ö–æ–¥–Ω—ã–π —Å–ª—É—á–∞–π–Ω—ã–π –≤–µ–∫—Ç–æ—Ä;\
$\vec{\xi}_{st} \sim \mathcal{N}(\vec{\mu}_{st}, \Sigma_{st})$ ‚Äî —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π –Ω–æ—Ä–º–∞–ª—å–Ω—ã–π —Å–ª—É—á–∞–π–Ω—ã–π –≤–µ–∫—Ç–æ—Ä;\
$\vec{\xi}_{st} = A \cdot \vec{\xi} + b$ ‚Äî –∞—Ñ–∏–Ω–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞;\
$\vec{\mu}_{st} = A \cdot \vec{\mu} + b$ ‚Äî –ø–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ–∂–∏–¥–∞–Ω–∏–π –∏–∑ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ;\

$\quad$\
–û—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –ø–µ—Ä–µ–≤–æ–¥—è—â–µ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –≤ –≤–µ–∫—Ç–æ—Ä —Å –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º–∏:\
$Q^{\intercal}$ ‚Äî –º–∞—Ç—Ä–∏—Ü–∞ –∏–∑ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã $\Sigma;$\
–ü—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ –≤—ã—Ä–∞–∂–µ–Ω–∏—è $Q \cdot \Sigma \cdot Q^{\intercal}$ –¥–æ–ª–∂–Ω–∞ –ø–æ–ª—É—á–∏—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–∞ $D$ –≤–∏–¥–∞: $\begin{pmatrix} \ldots & 0 \\ 0 & \ldots \end{pmatrix};$\
–ú–∞—Ç—Ä–∏—Ü–∞ $D$ —è–≤–ª—è–µ—Ç—Å—è –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ–π –∏–∑ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π $\Sigma^{-1},$\
–ø—Ä–∏ —ç—Ç–æ–º $\Sigma = Q \cdot D^{-1} \cdot Q^{\intercal},$\
–ü–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤ —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥:\
$p_{\xi,\eta}(x', y') = \frac{1}{2\pi \cdot \sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}} \cdot \exp\left(-\frac{1}{2} \cdot \left[  \begin{pmatrix} x' & y' \end{pmatrix} \cdot D^{-1} \cdot \begin{pmatrix} x' \\ y' \end{pmatrix} \right]\right)$\
$\quad$\
C–æ–≤–º–µ—Å—Ç–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞:\
–ü—É—Å—Ç—å $\vec{z} = A \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$ ‚Äî —Å–ª—É—á–∞–π–Ω—ã–π –≤–µ–∫—Ç–æ—Ä,\

## –ü—É–Ω–∫—Ç 1

–î–∞–Ω–æ: $ùõ≤_{\xi,\eta}(x,y) = ùê∂ ‚ãÖ \exp\left(-\frac{1}{2}\left(3x^2 + 3xy + 7y^2 + 9x + 17y + 13\right)\right)$

–î–ª—è –¥–∞–ª—å–Ω–µ–π—à–∏—Ö —à–∞–≥–æ–≤ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–¥–µ–ª–∏—Ç—å –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é —Ñ–æ—Ä–º—É:

$ 3x^2 + 3xy + 7y^2 + 9x + 17y + 13 = 3\left(x^2 + 2x + 1\right) + 7\left(y^2 + 2y + 1\right) + 3xy + 3x + 3y + 3 = \\ = 3\left(x + 1\right)^2 + 3\left(x + 1\right)\left(y + 1\right) + 7\left(y + 1\right)^2$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –º—ã –ø–æ–ª—É—á–∞–µ–º:

$ùõ≤_{\xi,\eta}(x,y) = ùê∂ ‚ãÖ \exp\left(-\frac{1}{2}\left[3\left(x + 1\right)^2 + 3\left(x + 1\right)\left(y + 1\right) + 7\left(y + 1\right)^2\right]\right)$

–ò–∑ —ç—Ç–æ–π –∑–∞–ø–∏—Å–∏ –º—ã –º–æ–∂–µ–º –ø–æ–ª—É—á–∏—Ç—å –º–∞—Ç—Ä–∏—Ü—É $Œ£^{-1}:$

$Œ£^{-1} =
\begin{pmatrix}
  3 & \frac{3}{2}\\
  \frac{3}{2} & 7
\end{pmatrix}$

–ü–æ–ª—É—á–∏–º –º–∞—Ç—Ä–∏—Ü—É $Œ£$:

$Œ£ = \frac{1}{\begin{vmatrix}Œ£^{-1}\end{vmatrix}} ‚ãÖ
\begin{pmatrix}
  7 & -\frac{3}{2}\\
  -\frac{3}{2} & 3
\end{pmatrix}
= \frac{4}{75} ‚ãÖ
\begin{pmatrix}
  7 & -\frac{3}{2}\\
  -\frac{3}{2} & 3
\end{pmatrix}$

–¢–∞–∫–∂–µ –∑–∞–ø–∏—à–µ–º –≤–µ–∫—Ç–æ—Ä –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ–∂–∏—Ç–∞–Ω–∏–π $\vecŒº_{Œæ,Œ∑}$:

$\vecŒº_{Œæ,Œ∑} =
\begin{pmatrix}
  -1 \\
  -1
\end{pmatrix}$

–ó–∞–ø–∏—à–µ–º –∏—Å–∫–æ–º—ã–µ $C, \mathbb{D}_Œæ$, $\mathbb{D}_\eta$, $\mathop{\mathrm{cov}}\nolimits(\xi, \eta)$ –∏ $\rho(\xi, \eta)$:

$
  ùê∂ = \frac{1}{2\pi \cdot \sqrt{\frac{4}{75}}} = \frac{5\sqrt{3}}{4\pi} \\
  \mathbb{D}_Œæ = \frac{4}{75} ‚ãÖ 7 = \frac{28}{75}\\
  \mathbb{D}_\eta = \frac{4}{75} ‚ãÖ 3 = \frac{4}{25}\\
  \mathop{\mathrm{cov}}\nolimits(\xi, \eta) = \frac{4}{75} ‚ãÖ \frac{-3}{2} = -\frac{2}{25}\\
  \rho(\xi, \eta) = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta)}
  {\sqrt{\mathbb{D}_Œæ ‚ãÖ \mathbb{D}_\eta}} = \frac{-\left(\frac{2}{25}\right)}{\sqrt{\left(\frac{28}{75}\right) ‚ãÖ \left(\frac{4}{25}\right)}} = - \frac{\sqrt{21}}{14}\\
$

–î–∞–ª–µ–µ –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $ùõ≤_{\xi,\eta}$:

```r
P_xi_eta <- function(x, y) {
  5 * sqrt(3) / (4 * pi) * exp(-0.5 * (3 * (x + 1)^2 + 3 * (x + 1) * (y + 1) + 7 * (y + 1)^2))
}

x <- seq(-4, 4, length = 50)
y <- seq(-4, 4, length = 50)
z <- outer(x, y, P_xi_eta)

persp(x, y, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 0, phi = 0)
persp(x, y, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 90, phi = 90)
persp(x, y, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 45, phi = 45)
```

## –ü—É–Ω–∫—Ç 3

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø–µ—Ä–µ–≤–¥–µ—Ç –Ω–∞—à —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ª—É—á–∞–π–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –≤ –≤–µ–∫—Ç–æ—Ä —Å –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º–∏. –î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –º–∞—Ç—Ä–∏—Ü—ã $Œ£^{-1}. –°–¥–µ–ª–∞–µ–º —ç—Ç–æ —Å –ø–æ–º–æ—â—å—é –∫–æ–¥–∞ –Ω–∞ —è–∑—ã–∫–µ R

$Œ£^{-1}_{new} = Q \cdot Œ£^{-1} \cdot Q^{-1} =

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≤–∑—è–≤ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏–∑ $Œ£^{-1}_{new}$ –ø–æ–ª—É—á–∏–º —Å–ª–µ–¥—É—é—â—É—é —Ñ–æ—Ä–º—É–ª—É –¥–ª—è –ø–ª–æ—Ç–Ω—Ç—Å—Ç–∏:

$ùõ≤'_{\xi,\eta}(x',y') = ùê∂' ‚ãÖ \exp\left(-\frac{1}{2}\left(2.5x'^2 + 8y'^2\right)\right)$
–í—ã—á–∏—Å–ª–∏–º $C'$:

$C' = \frac{1}{2\pi \cdot \sqrt{|Œ£_{new}|}} = \frac{1}{2\pi \cdot \sqrt{\frac{4}{75}}} = \frac{1}{2\pi \cdot \sqrt{\frac{4}{75}}} = \frac{5\sqrt{3}}{4\pi}$
–ò—Ç–æ–≥–æ –ø–æ–ª—É—á–∏–º:
$ùõ≤'_{\xi,\eta}(x',y') = \frac{5\sqrt{3}}{4\pi} ‚ãÖ \exp\left(-\frac{1}{2}\left(2.5x'^2 + 8y'^2\right)\right)$

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

```r
P_xi_eta <- function(x, y) {
  5 * sqrt(3) / (4 * pi) * exp(-0.5 * (2.5 * x^2 + 7.5 * y^2))
}

x <- seq(-4, 4, length = 50)
y <- seq(-4, 4, length = 50)
z <- outer(x, y, P_xi_eta)

persp(x, y, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 0, phi = 0)
persp(x, y, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 90, phi = 90)
persp(x, y, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 45, phi = 45)
```

##–ü—É–Ω–∫—Ç 4
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Å–æ–≤–º–µ—Å—Ç–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ $\left(-3\xi -5\eta, -\xi + \eta\right)^T$ –∏ –∑–∞–ø–∏—Å–∞—Ç—å –µ–≥–æ –ø–ª–æ—Ç–Ω–æ—Å—Ç—å. –¢–∞–∫–∂–µ –Ω–∞–¥–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

–ü—É—Å—Ç—å $\vec{t}=
\begin{pmatrix}
  -3\xi -5\eta \\
  -\xi + \eta
\end{pmatrix}
= A \cdot
\begin{pmatrix}
  \xi \\
  \eta
\end{pmatrix},$\
—Ç–æ–≥–¥–∞ $\vec{t} \sim \mathcal{N}(A \cdot \vec{\mu}, A \cdot \Sigma \cdot A^{\intercal})$, –≥–¥–µ –º–∞—Ç—Ä–∏—Ü–∞ $A = \begin{pmatrix}
  -3 & -5 \\
  -1 & 1
\end{pmatrix}$\
–û–±–æ–∑–Ω–∞—á–∏–º $A \cdot \vec{\mu} = \mathbb{E}[\vec{z}] = \vec{\mu}_{\vec{z}}$ –∏ $A \cdot \Sigma \cdot A^{\intercal} = \Sigma_{\vec{t}}$\
–ü—Ä–æ–∏–∑–≤–µ–¥–µ–º —Ä–∞—Å—á–µ—Ç—ã:

```r
mu <- matrix(c(-1, -1), ncol = 1)
Sigma <- matrix(c(28/75, -2/25, -2/25, 4/25), nrow = 2)
A <- matrix(c(-3, -1, -5, 1), nrow = 2)

mu_t <- A %*% mu
Sigma_t <- A %*% Sigma %*% t(A)

det_Sigma_z <- (Sigma_z[1] * Sigma_z[4] - (Sigma_z[2] * Sigma_z[3]))

cat("mu_t:","\n")
prmatrix(mu_t)

cat("Sigma_t ‚ãÖ 75/4:","\n")
prmatrix(Sigma_t * 75/4)

cat("det_Sigma_z ‚ãÖ 75/4:", det_Sigma_z * 75/4)
```

$Œ£_{\vec{t}} = \frac{4}{75} \cdot \begin{pmatrix}
  93 & 3 \\

\end{pmatrix} \\
\vec{\mu}_\vec{t} =
\begin{pmatrix}
  8 \\

\end{pmatrix} \\
Œ£_{\vec{t}}^{-1} = \frac{75}{256} \cdot \begin{pmatrix}

  -3 & 93
\end{pmatrix}$

–ó–Ω–∞—è –¥–∞–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏:
$ùõ≤_{\vec{t}}(\vec{t}) = \frac{8\sqrt{3}}{15\pi} ‚ãÖ \exp\left(-\frac{1}{2} \cdot \frac{75}{256}\left[13\left(Œæ - 8\right)^2 -6\left(Œæ - 8\right)Œ∑ + 93Œ∑^2\right]\right)$

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:


```r
P_z <- function(xi, eta) {
  8 * sqrt(3) / (15 * pi) * exp(-0.5 * (75/256) * (13 * (xi - 8)^2 - 6 * (xi - 8) * eta + 93 * eta^2))
}

xi <- seq(0, 15, length = 50)
eta <- seq(-7, 8, length = 50)
z <- outer(xi, eta, P_z)

persp(xi, eta, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 0, phi = 0)
persp(xi, eta, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 90, phi = 90)
persp(xi, eta, z, col = "lightblue", border = "black", ticktype = "detailed", theta = 45, phi = 45)
```

–ù–∞–π–¥–µ–º —É—Å–ª–æ–≤–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $\xi$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ $\eta$.
–î–ª—è –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —É—Å–ª–æ–≤–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ç–∞–∫–∂–µ —è–≤–ª—è–µ—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º. –£—Å–ª–æ–≤–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $\xi | \eta = y$ –∏–º–µ–µ—Ç —Å–ª–µ–¥—É—é—â—É—é —Ñ–æ—Ä–º—É:\
$\xi | \eta = y \sim \mathcal{N}(\mathbb{E}[\xi|\eta], \mathbb{D}[\xi|\eta])$, –≥–¥–µ\
$\mathbb{E}[\xi|\eta] = \mu_{\xi} + \frac{\sigma_{\xi\eta}}{\sigma_{\eta}^2} \cdot (y - \mu_{\eta}),$\
$\mathbb{D}[\xi|\eta] = \sigma_{\xi}^2 - \frac{\sigma_{\xi\eta}^2}{\sigma_{\eta}^2}.$\
–í—ã—á–∏—Å–ª–∏–º $\mathbb{E}[\xi|\eta]$ –∏ $\mathbb{D}[\xi|\eta]:$\
$\mathbb{E}[\xi|\eta] = -1 + \frac{-\frac{2}{25}}{\frac{4}{25}} \cdot (y_{0} + 1) = -1 -\frac{1}{2} \cdot (y_{0} + 1) = -1 -\frac{1}{2} \cdot y_{0} -\frac{1}{2} = -\frac{3}{2} -\frac{1}{2} \cdot y_{0};$\
$\mathbb{D}[\xi|\eta] = \frac{28}{75} - \frac{(-\frac{2}{25})^2}{\frac{4}{25}} = \frac{28}{75} - \frac{\frac{4}{625}}{\frac{4}{25}} = \frac{28}{75} - \frac{1}{25} = \frac{25}{75} = \frac{1}{3}.$\
–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —É—Å–ª–æ–≤–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $\xi | \eta = y_{0}$ –∏–º–µ–µ—Ç –≤–∏–¥:\
$\xi | \eta = y_{0} \sim \mathcal{N}(-\frac{3}{2} -\frac{1}{2} \cdot y_{0}, \frac{1}{3}),$
–∞ —Ñ—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥:\

\exp\left( -\frac{(x - \mu_{\xi|\eta = y_{0}})^2}{2\sigma_{\xi|\eta}^2} \right);$\
